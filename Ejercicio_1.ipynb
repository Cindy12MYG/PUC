```python
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk # load skimage libraries
from skimage import io

file_path = '/content/drive/MyDrive/Github/monet.jpg'
I1 = sk.io.imread(file_path) # read image using imread function
I1.shape # 3 matrices : couleur
```

```python
file_path = '/content/drive/MyDrive/Github/matches.tif'
I2 = sk.io.imread(file_path) # read image using imread function
I2.shape # 1 matrice : gris
```

```python
fig = plt.figure(figsize=(14, 6))
plt.subplot(121) # 1 ligne 2 colonne, image 1
sk.io.imshow(I1) # visualize image using imshow function
plt.title('First image')

plt.subplot(122) # 1 ligne 2 colonne, image 2
sk.io.imshow(I2) # visualize image using imshow function
plt.title('Second image')
plt.show()
```

```python
IG = sk.color.rgb2gray(I1)
IG.shape # 1 matrice : gris
IG.dtype
```

```python
fig = plt.figure(figsize=(10,6))
sk.io.imshow(IG) # visualize image using imshow function
sk.io.show()
```

```python
import matplotlib.pyplot as plt
fig,axes = plt.subplots(1,2,figsize=(12,7))
ax = axes.ravel() # convert axes to a 1d array
im1 = ax[0].imshow(IG,cmap=plt.cm.gray) # visualize image using imshow function using a grayscale colormap # cmap=plt.cm.gray met l'échelle en gris mais l'image n'est pas grise car de 0 à 1
ax[0].set_title('grayscale image with grayscale colormap') # set figure title
fig.colorbar(im1,ax=ax[0],orientation='horizontal') # visualize colorbar
im2 = ax[1].imshow(IG,cmap=plt.cm.autumn) # visualize image using imshow function using a grayscale colormap ici filtre jet, on peut remplacer par winter par ex ou autumn
ax[1].set_title('grayscale image with autumn colormap') # set figure title
fig.colorbar(im2,ax=ax[1],orientation='horizontal') # visualize colorbar
sk.io.show()
```

```python
# reshape the uint8 image as 1d array: (= un vecteur)
v = IG.reshape(IG.shape[0]*IG.shape[1],1)

fig = plt.figure(figsize=(14, 6))
plt.hist(v, bins=256,density=False) # 256 = nbr de rectangles, le max de valeurs qu'il prend
plt.show()
```
